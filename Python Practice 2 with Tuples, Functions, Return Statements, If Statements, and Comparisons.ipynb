```python
#Tuples
#These are type of data structure, like container that stores different values
#Tuples similar to lists but have few key differences
#one of most common examples of tuples is coordinates
#whenever we create tuple, need to have open and closed parantheses
coordinates = (4,5)  #4 and 5 are values stored in coordinates which is tuple
#unique thing about tuple is that it's inmutable, meaning it can't be changed or modified
# if we put coordinates[1] = 10 (which would set the 5 in index 1 to 10), we get error since tuple cannot be changed
print(coordinates[0]) #use [] to access values in tuple--> prints 4 since 4 is in index 0
print(coordinates[1]) #prints 5 since 5 is in index position 1
#one difference between lists and tuples is lists use [] while tuples use () -> also list can modify values but tuples can't
# coordinates = [4,5] puts values into list instead of tuple since it has [] instead of ()
#tuples used for data that will never change, so coordinates is good use of them (they don't change, just get added to others) 
coordinates = [(4,5), (6,7), (80,34)]   #this is list with many tuples inside that cannot be modified
#lists used majority of the time while tuples are more niche and really only used for data that does not change
```

```python
#Functions
#These are collection of code that performs specific task--> just call on function to perform task
#Let's make function whose task is to say hi to user
#when making function, need to use keyword def (want to write functions in lowercase and use _ to separate words)
def say_hi(): #this tells python that all code after this line will be inside this function
    #in order to write code inside this function, need to indent (anything not indented not considered part of function)
    print("Hello User")  #says hello to user as aprt of function
    #if you run program as is, nothing will happen since you need to call on function
say_hi() #to call function, just type function's name and () after --> this prints Hello User since sayhi function is called
print("Top")                #     Top
say_hi() #output looks like this: Hello User
print("Bottom")             #     Bottom
#this is because python prints Top on first line, then executes code in say_hi function and then prints Bottom on last line 
#we can pass information into function which is called a parameter
def say_hi(name):   #changes say_hi function so that it has to take a parameter
    print("Hello " + name + "! ")    #prints Hello Mike! since it says hello to whatever name got passed as parameter
say_hi("Mike")  #passes parameter "Mike" into say_hi function
say_hi("Steve") #passes parameter "Steve" into say_hi function
#now let's try to make function taken more than one parameter
def say_hi(name, age):   #now, two parameters passed into say_hi function which are name and age
    print("Hello " + name + ", you are " + age + ". ")   #looks at parameters in code and prints using those parameters
say_hi("Mike","35")  #Mike is name parameter and 35 is age parameter so it prints Hello Mike, you are 35.
say_hi("Steve", "70")  #Steve is name parameter and 70 is age parameter so it prints Hello Steve, you are 70.
#can pass string or number into function but need to convert number into string first before printing
#can also pass booleans and arrays into functions
def say_hi(name, age):
    print("Hello " + name + ", you are " + str(age) + ". ")  #since age parameter is number now, use str() to convert to string 
say_hi("Mike", 35)  #this time, 35 is number so need to convert to string --> prints Hello Mike, you are 35 years old.
say_hi("Steve", 70)  #this time, 70 is number so need to convert to string --> prints Hello Steve, you are 70 years old.
```

```python
#Return Statement
#use this to get back information from function after calling on it
#return keyword allows python to return information from function
#Let's make function that takes the cube of a number
def cube(num):  #creates function called cube with num as parameter
    num*num*num  #when called, cube function will multiply number parameter by itself three times, essentially cubing it
cube(3) #this won't do anything so nothing gets printed
print(cube(3))  #this prints None since function is doing nothing without return statement
#different from previous case where say_hi function had print() code included in function
def cube(num):
    return num*num*num  #return means python executes code in function and then realizes it needs to return value of num*num*num
    print("code")  #this will never get printed because return means breaking out of function which python wll do 
print("code")  #this will get printed because it is outside of function and not affected by return statement
cube(3)    
print(cube(3))   #now, when we call cube function, it returns num*num*num which is 27 in this case so it prints 27
print(cube(4))  #prints 64 since python returns num*num*num value inside cube function
result = cube(4)  #variable called result will store value that gets returned from cube function (not gonna store cube(4))
print(result)  #prints 64 since that is value that it's storing
#we use parameter to give information to function and the return statement gives back information from function
#Can return any data type you want like string, boolean or array
```

```python
#If Statements and Comparisons
#instead of using boolean values, we can compare different values
def max_num(num1, num2, num3):  #creates function called max_num which needs three number parameters and returns biggest number
#to find out which number is biggest, need if statements    
    if num1 >= num2 and num1 >= num3:    #condition is if num1 is greater than/equal to num2 & it's greater than/equal to num3
#when we compare these two numbers, we end up with true or false value
#num1 being greater than or equal to num2 is either true or false, making it a boolean value (>= is called comparison operator)
        return num1  #if both conditions end up being true, then num1 is biggest number, so function returns num1
#but what about if num2 or num3 is the biggest number
    elif num2 >= num1 and num2 >= num3:  #condition is if num2 is greater than/equal to num1 & it's greater than/equal to num3
        return num2   #if both conditions are true, then num2 is biggest number, so function returns num2
    else: #don't need conditions because if num1 and num2 aren't biggest, then num3 has to be biggest
        return num3   #if all other conditions false & num1 & num2 aren't biggest, then it must be num3 so function returns num3
print(max_num(3,4,5))  #calls max_num function & passes 3,4,5 as number parameters, then prints 5 which is biggest number
print(max_num(3,40,5)) #calls max_num function & passes 3,4,5 as number parameters, then prints 40 which is biggest number
print(max_num(300,40,5)) #calls max_num function & passes 3,4,5 as number parameters, then prints 300 which is biggest number
#These comparisons are common way to make if statements (can compare numbers, strings, and booleans)
#Other comparison operators include == which checks to see if something is equal to something else
#Also, there's <, <=, >, >= and != which means not equal to
```
